```python

#solution:(5)


import numpy as np
x=[1,2,3,4,5] 
for i in x:
    if i%2==0:
        x.remove(i)
print(x)    

y=[5,6,7,8,9]
for j in y:
    if j%2==0:
        y.remove(j)
print(y)       
merge=np.concatenate((x,y))
print ("output with duplicate values =", merge)
convert_set=set(merge)
result=list(convert_set)
print ("Expected output with unique values =", result)
```
